## Plotting a line

```python
import matplotlib.pyplot as plt 

x = [1,2,3]
y = [2,4,1]

# plot x and y and add labels and title
plt.plot(x, y) 
plt.xlabel('x - axis') 
plt.ylabel('y - axis')
plt.title('My first graph!') 
# finally show the graph
plt.show() 
```

## Plotting two or more lines on same plot

```python
import matplotlib.pyplot as plt 

x1 = [1,2,3] 
y1 = [2,4,1] 
plt.plot(x1, y1, label = "line 1")  # 给两个曲线也加上label

x2 = [1,2,3] 
y2 = [4,1,3] 
plt.plot(x2, y2, label = "line 2") 

plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.title('Two lines on same graph!') 

plt.legend()     # 显示line1 和 line2。右上角那个box就叫legend
plt.show()
```

## Customization of Plots

```python
x = [1,2,3,4,5,6] 
y = [2,4,1,5,2,6] 

# plot可以加参数，来设置曲线的样式
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 2, 
         marker='o', markerfacecolor='blue', markersize=8) 
  
# setting x and y axis range 
plt.ylim(1,8) 
plt.xlim(1,8) 
  
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.title('Some cool customizations!') 
  
# function to show the plot 
plt.show() 
```

## Bar Chart

```python
import matplotlib.pyplot as plt 

left = [1, 2, 3, 4, 5] 
height = [10, 24, 36, 40, 5] 
  
# labels for bars 
tick_label = ['one', 'two', 'three', 'four', 'five'] 
# plotting a bar chart 
plt.bar(left, height, tick_label = tick_label, 
        width = 0.8, color = ['red', 'green', 'blue']) 
  
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.title('My bar chart!') 
  
plt.show() 
```
